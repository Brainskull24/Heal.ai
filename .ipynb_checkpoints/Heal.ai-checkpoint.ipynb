```python
increase_count = 70

# Identify indices where 'rarely' is present
rarely_indices = df[df['SOCIAL ENGAGEMENT'] == 'rarely'].index

# Randomly select indices from 'rarely' and update them to 'frequently' and 'occasionally'
selected_indices = np.random.choice(rarely_indices, size=increase_count * 2, replace=False)

# Update the 'new_social_engagement' column
df.loc[selected_indices[:increase_count], 'SOCIAL ENGAGEMENT'] = 'frequently'
df.loc[selected_indices[increase_count:], 'SOCIAL ENGAGEMENT'] = 'occasionally'

# Update the counts in the new column
new_counts = df['SOCIAL ENGAGEMENT'].value_counts()
```

```python
for index, row in df.iterrows():
    if row['OCCUPATION'] == 'Teacher' and row['AGE'] <= 20:
        df.at[index, 'OCCUPATION'] = 'Student'

for index, row in df.iterrows():
    if row['OCCUPATION'] == 'Student' and row['AGE'] >= 30:
        df.at[index, 'OCCUPATION'] = 'Teacher'
```

```python
import numpy as np

# Assign scale values to the 'FREQUENCY' attribute based on percentages
frequency_values = np.random.choice([1, 2, 3, 4], size=len(df), p=[0.23, 0.31, 0.19, 0.27])
df['FREQUENCY'] = frequency_values

# Define the pattern for assigning scale values based on the frequency
def assign_scale_values(freq_value):
    if freq_value == 1:
        return np.random.randint(1, 3)  # Assign values between 1 and 2
    elif freq_value == 2:
        return np.random.randint(1, 4)  # Assign values between 1 and 3
    elif freq_value == 3:
        return np.random.randint(2, 5)  # Assign values between 2 and 4
    elif freq_value == 4:
        return np.random.randint(3, 5)  # Assign values between 3 and 4

# Apply the pattern to assign scale values to each attribute
df['Anxiety/Stress Away From Phone'] = df['FREQUENCY'].apply(assign_scale_values)
df['Feeling Without Phone'] = df['FREQUENCY'].apply(assign_scale_values)
df['SOCIAL ENGAGEMENT'] = df['FREQUENCY'].apply(lambda x: 1 if x == 4 else np.random.randint(1, 5))
df['Reduction Success'] = df['FREQUENCY'].apply(assign_scale_values)
df['Focus Ability'] = df['FREQUENCY'].apply(assign_scale_values)
df['Phone Impact on Lifestyle'] = df['FREQUENCY'].apply(assign_scale_values)
df['sleep time'] = df['FREQUENCY'].apply(assign_scale_values)

# Display the updated DataFrame
print(df.head())
```

```python
# Define attribute names
attribute_names = ['FREQUENCY', 'Anxiety/Stress Away From Phone', 'Feeling Without Phone', 
                   'SOCIAL ENGAGEMENT', 'Reduction Success', 'Focus Ability', 
                   'Phone Impact on Lifestyle', 'sleep time']

# Create a new column 'SUM' containing the sum of all values in each row
df['SUM'] = df[attribute_names].sum(axis=1)

df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define attribute names
attributes = ['FREQUENCY', 'Anxiety/Stress Away From Phone', 'Feeling Without Phone',
              'SOCIAL ENGAGEMENT', 'Reduction Success', 'Focus Ability',
              'Phone Impact on Lifestyle', 'sleep time']

# Compute descriptive statistics
mean_values = df[attributes].mean()
median_values = df[attributes].median()
std_deviation = df[attributes].std()
min_values = df[attributes].min()
max_values = df[attributes].max()
quantiles = df[attributes].quantile([0.25, 0.5, 0.75])
skewness = df[attributes].skew()
kurtosis = df[attributes].kurt()
variance = np.var(df[attributes], axis=0, ddof=1)

# Plot histograms
for column in attributes:
    plt.hist(df[column], bins=10)
    plt.title(f'Histogram of {column}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

# Compute correlation matrix
correlation_matrix = df[attributes].corr()

# Print descriptive statistics
print("Mean Values:")
print(mean_values)
print("\nMedian Values:")
print(median_values)
print("\nMinimum Values:")
print(min_values)
print("\nMaximum Values:")
print(max_values)
print("\nQuantiles:")
print(quantiles)
print("\nCorrelation Matrix:")
print(correlation_matrix)
print("\nSkewness:")
print(skewness)
print("\nKurtosis:")
print(kurtosis)
print("\nStandard Deviation:")
print(std_deviation)
print("\nVariance:")
print(variance)
```

```python
def cronbach_alpha(dataset):
    items_count = dataset.shape[1]
    variances_sum = np.var(dataset, axis=0, ddof=1).sum()
    total_var = np.var(np.sum(dataset, axis=1), ddof=1)
    alpha = (items_count / (items_count - 1)) * (1 - (variances_sum / total_var))
    return alpha

questions_data = sdf[['FREQUENCY', 'Anxiety/Stress Away From Phone', 'Feeling Without Phone',
              'SOCIAL ENGAGEMENT', 'Reduction Success', 'Focus Ability',
              'Phone Impact on Lifestyle', 'sleep time']]
data_array = questions_data.to_numpy()
print("Cronbach's Alpha:", cronbach_alpha(data_array))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Fill NaN values with 0
sdf.fillna(0, inplace=True)

# Define function to categorize SAS scores
def categorize_sas(score):
    if 1 <= score <= 18:
        return 'NOT ADDICTED'
    else:
        return 'ADDICTED'

# Apply categorization to create 'SAS RESULT' column
sdf['SAS RESULT'] = sdf['SUM'].apply(categorize_sas)

# Define features (X) and target (y)
X = sdf.drop(['SUM', 'SAS RESULT', 'NAME', 'AGE', 'GENDER','PHONE SCREEN TIME'], axis=1)
y = sdf['SAS RESULT']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the SVM classifier
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train, y_train)

# Make predictions on the test set
predictions_svm = svm_classifier.predict(X_test)

# Calculate accuracy
accuracy_svm = accuracy_score(y_test, predictions_svm)

# Generate classification report
classification_rep_svm = classification_report(y_test, predictions_svm)

# Sample row for prediction
sample_row = [[2, 1, 4, 3, 2, 3, 1, 4]]

# Predict SAS result for the sample row
prediction = svm_classifier.predict(sample_row)

# Display results
print("Predicted Depression Level:", prediction)
print("SVM Accuracy:", accuracy_svm)
print("SVM Classification Report:")
print(classification_rep_svm)
```

```python
svm_classifier = SVC(kernel='poly', random_state=42)
svm_classifier.fit(X_train, y_train)
predictions_svm = svm_classifier.predict(X_test)

accuracy_svm = accuracy_score(y_test, predictions_svm)
classification_rep_svm = classification_report(y_test, predictions_svm)
sample_prediction_svm = svm_classifier.predict(sample_row)
print(sample_row)
print("SVM Predicted Depression Level:", sample_prediction_svm)
print("SVM Accuracy:", accuracy_svm)
print("SVM Classification Report:")
print(classification_rep_svm)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test, predictions_svm)

sensitivity = []
specificity = []
for i in range(len(np.unique(y_test))):
    TP = conf_matrix[i, i]
    FN = np.sum(conf_matrix[i, :]) - TP
    FP = np.sum(conf_matrix[:, i]) - TP
    TN = np.sum(conf_matrix) - TP - FN - FP

    sensitivity.append(TP / (TP + FN))
    specificity.append(TN / (TN + FP))

for i, class_label in enumerate(np.unique(y_test)):
    print(f"Class: {class_label}")
    print(f"Sensitivity: {sensitivity[i]:.2f}")
    print(f"Specificity: {specificity[i]:.2f}")
    print()
```

```python
def cronbach_alpha(dataset):
    items_count = dataset.shape[1]
    variances_sum = np.var(dataset, axis=0, ddof=1).sum()
    total_var = np.var(np.sum(dataset, axis=1), ddof=1)
    alpha = (items_count / (items_count - 1)) * (1 - (variances_sum / total_var))
    return alpha

questions_data = data[['phq1', 'phq2', 'phq3', 'phq4', 'phq5', 'phq6', 'phq7', 'phq8', 'phq9']]
data_array = questions_data.to_numpy()
print("Cronbach's Alpha:", cronbach_alpha(data_array))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.svm import SVC

# Replace missing values with zeros
data.fillna(0, inplace=True)
def categorize_sum(score):
    if 1 <= score <= 4:
        return 'Minimal depression'
    elif 5 <= score <= 9:
        return 'Mild depression'
    elif 10 <= score <= 14:
        return 'Moderate depression'
    elif 15 <= score <= 19:
        return 'Moderately severe depression'
    elif 20 <= score <= 26:
        return 'Severe depression'
    else:
        return 'No depression'

data['Depression_Severity'] = data['sum'].apply(categorize_sum)
X = data.drop(['sum', 'Depression_Severity'], axis=1)
y = data['Depression_Severity'] 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
svm_classifier = SVC(kernel='linear', random_state=42)
svm_classifier.fit(X_train, y_train)

predictions_svm = svm_classifier.predict(X_test)
accuracy_svm = accuracy_score(y_test, predictions_svm)
classification_rep_svm = classification_report(y_test, predictions_svm)
sample_row = [[2, 3, 2, 1, 4, 3, 2, 3, 1]]
prediction = svm_classifier.predict(sample_row)

print("Predicted Depression Level:", prediction)
print("SVM Accuracy:", accuracy_svm)
print("SVM Classification Report:")
print(classification_rep_svm)
```

```python
svm_classifier = SVC(kernel='poly', random_state=42)
svm_classifier.fit(X_train, y_train)
predictions_svm = svm_classifier.predict(X_test)

accuracy_svm = accuracy_score(y_test, predictions_svm)
classification_rep_svm = classification_report(y_test, predictions_svm)
sample_prediction_svm = svm_classifier.predict(sample_row)
print(sample_row)
print("SVM Predicted Depression Level:", sample_prediction_svm)
print("SVM Accuracy:", accuracy_svm)
print("SVM Classification Report:")
print(classification_rep_svm)
```
